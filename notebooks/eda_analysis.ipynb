### Veri Seti İstatistiksel Özeti
Bu aşamada veri setimizdeki sayısal değişkenler üzerinde istatistiksel bir özet çıkardık. Çıktılar arasında her bir değişkenin **ortalama (mean)**, **standart sapma (std)**, **minimum (min)** ve **maksimum (max)** değerleri gibi istatistiksel bilgileri yer alıyor. Ayrıca çeyrek değerler olan **25%**, **50% (median)** ve **75%** yüzdeliklerini de görebiliyoruz.

Bu özet, veri setimizdeki değişkenlerin genel dağılımını anlamamıza yardımcı oldu. Özellikle, her bir değişkenin **min ve max** değerlerini analiz ederek uç değerleri veya olası anormallikleri tespit edebiliriz.

Bir sonraki adımda, sayısal değişkenlerin dağılımlarını daha iyi anlamak için histogramlarla görselleştirme yapacağız.


```python
# Sayısal sütunları seçelim
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Histogramlar çizelim
fig, axes = plt.subplots(nrows=8, ncols=7, figsize=(20, 20))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

for i, column in enumerate(numeric_columns):
    row = i // 7
    col = i % 7
    ax = axes[row, col]
    df[column].hist(ax=ax, bins=15, edgecolor='black')
    ax.set_title(column)
    ax.set_xlabel('Value')  # Eksen adı, burada genelleştirilmiş
    ax.set_ylabel('Frequency')  # Eksen adı, burada genelleştirilmiş

# Kalan boş subplotları kaldır
for j in range(len(numeric_columns), 8 * 7):
    fig.delaxes(axes.flatten()[j])

plt.tight_layout()
plt.show()
```

## Histogram Analizi

Aşağıdaki grafiklerde, veri kümesindeki sayısal sütunlar için histogramlar gösterilmektedir. Histogramlar, her bir sütunun dağılımını ve frekansını görsel olarak incelememize yardımcı olur.

### Histogramlar

- **X Eksenleri**: Değerler (Value) - Her sütun için gösterilen değerlerin aralığını temsil eder.
- **Y Eksenleri**: Frekans (Frequency) - Belirli değer aralıklarında kaç gözlem bulunduğunu gösterir.

#### Grafiklerin Açıklamaları

1. **Ability**: Oyuncunun yetenek puanlarının dağılımı.
2. **Potential**: Oyuncunun potansiyel puanlarının dağılımı.
3. **Positions**: Oyuncunun oynayabileceği pozisyon sayısının dağılımı.
4. **Caps / Goals**: Oyuncunun uluslararası maçlarındaki gol sayılarının dağılımı.
5. **Foot**: Sağ veya sol ayak tercihlerinin dağılımı.
6. **Height**: Oyuncuların boylarının dağılımı (genellikle metre cinsinden).
7. **Weight**: Oyuncuların kilolarının dağılımı (genellikle kilogram cinsinden).
8. **Aerial Reach**: Oyuncunun hava toplarında ne kadar etkili olduğunu gösteren puanların dağılımı.

Diğer histogramlar da benzer şekilde, oyuncuların çeşitli fiziksel ve teknik özelliklerinin dağılımını gösterir.

### Notlar

- Histogramların her biri, veri kümesindeki farklı bir sayısal özelliği temsil eder.
- X eksenindeki değerler, veri kümesindeki özelliklerin aralığını belirtirken; Y ekseni, bu değerlerin kaç kez tekrarlandığını (frekansını) gösterir.
- Grafiklerde görülen dağılımlar, veri kümesinin genel yapısını anlamamıza yardımcı olur ve bu özelliklerin çeşitli analizlerde nasıl kullanılabileceği konusunda bilgi verir.


```python
# ID ve gereksiz sütunları çıkararak korelasyon matrisini hesapla
corr_matrix = df.drop(columns=["id"]).corr().abs()

# Korelasyon matrisinin üst üçgenini al
upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# En yüksek ve en düşük korelasyonları bul
most_correlated = upper_triangle.stack().sort_values(ascending=False)

# En yüksek 10 korelasyon
print("En yüksek 10 korelasyon:")
print(most_correlated.head(10))

# En düşük 10 korelasyon
print("\nEn düşük 10 korelasyon:")
print(most_correlated.tail(10))

# Tüm korelasyonları göster
print("\nTüm korelasyonlar:")
print(most_correlated)
```
