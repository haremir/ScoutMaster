### Veri Seti Ä°statistiksel Ã–zeti
Bu aÅŸamada veri setimizdeki sayÄ±sal deÄŸiÅŸkenler Ã¼zerinde istatistiksel bir Ã¶zet Ã§Ä±kardÄ±k. Ã‡Ä±ktÄ±lar arasÄ±nda her bir deÄŸiÅŸkenin **ortalama (mean)**, **standart sapma (std)**, **minimum (min)** ve **maksimum (max)** deÄŸerleri gibi istatistiksel bilgileri yer alÄ±yor. AyrÄ±ca Ã§eyrek deÄŸerler olan **25%**, **50% (median)** ve **75%** yÃ¼zdeliklerini de gÃ¶rebiliyoruz.

Bu Ã¶zet, veri setimizdeki deÄŸiÅŸkenlerin genel daÄŸÄ±lÄ±mÄ±nÄ± anlamamÄ±za yardÄ±mcÄ± oldu. Ã–zellikle, her bir deÄŸiÅŸkenin **min ve max** deÄŸerlerini analiz ederek uÃ§ deÄŸerleri veya olasÄ± anormallikleri tespit edebiliriz.

Bir sonraki adÄ±mda, sayÄ±sal deÄŸiÅŸkenlerin daÄŸÄ±lÄ±mlarÄ±nÄ± daha iyi anlamak iÃ§in histogramlarla gÃ¶rselleÅŸtirme yapacaÄŸÄ±z.


```python
# SayÄ±sal sÃ¼tunlarÄ± seÃ§elim
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Histogramlar Ã§izelim
fig, axes = plt.subplots(nrows=8, ncols=7, figsize=(20, 20))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

for i, column in enumerate(numeric_columns):
    row = i // 7
    col = i % 7
    ax = axes[row, col]
    df[column].hist(ax=ax, bins=15, edgecolor='black')
    ax.set_title(column)
    ax.set_xlabel('Value')  # Eksen adÄ±, burada genelleÅŸtirilmiÅŸ
    ax.set_ylabel('Frequency')  # Eksen adÄ±, burada genelleÅŸtirilmiÅŸ

# Kalan boÅŸ subplotlarÄ± kaldÄ±r
for j in range(len(numeric_columns), 8 * 7):
    fig.delaxes(axes.flatten()[j])

plt.tight_layout()
plt.show()
```

## Histogram Analizi

AÅŸaÄŸÄ±daki grafiklerde, veri kÃ¼mesindeki sayÄ±sal sÃ¼tunlar iÃ§in histogramlar gÃ¶sterilmektedir. Histogramlar, her bir sÃ¼tunun daÄŸÄ±lÄ±mÄ±nÄ± ve frekansÄ±nÄ± gÃ¶rsel olarak incelememize yardÄ±mcÄ± olur.

### Histogramlar

- **X Eksenleri**: DeÄŸerler (Value) - Her sÃ¼tun iÃ§in gÃ¶sterilen deÄŸerlerin aralÄ±ÄŸÄ±nÄ± temsil eder.
- **Y Eksenleri**: Frekans (Frequency) - Belirli deÄŸer aralÄ±klarÄ±nda kaÃ§ gÃ¶zlem bulunduÄŸunu gÃ¶sterir.

#### Grafiklerin AÃ§Ä±klamalarÄ±

1. **Ability**: Oyuncunun yetenek puanlarÄ±nÄ±n daÄŸÄ±lÄ±mÄ±.
2. **Potential**: Oyuncunun potansiyel puanlarÄ±nÄ±n daÄŸÄ±lÄ±mÄ±.
3. **Positions**: Oyuncunun oynayabileceÄŸi pozisyon sayÄ±sÄ±nÄ±n daÄŸÄ±lÄ±mÄ±.
4. **Caps / Goals**: Oyuncunun uluslararasÄ± maÃ§larÄ±ndaki gol sayÄ±larÄ±nÄ±n daÄŸÄ±lÄ±mÄ±.
5. **Foot**: SaÄŸ veya sol ayak tercihlerinin daÄŸÄ±lÄ±mÄ±.
6. **Height**: OyuncularÄ±n boylarÄ±nÄ±n daÄŸÄ±lÄ±mÄ± (genellikle metre cinsinden).
7. **Weight**: OyuncularÄ±n kilolarÄ±nÄ±n daÄŸÄ±lÄ±mÄ± (genellikle kilogram cinsinden).
8. **Aerial Reach**: Oyuncunun hava toplarÄ±nda ne kadar etkili olduÄŸunu gÃ¶steren puanlarÄ±n daÄŸÄ±lÄ±mÄ±.

DiÄŸer histogramlar da benzer ÅŸekilde, oyuncularÄ±n Ã§eÅŸitli fiziksel ve teknik Ã¶zelliklerinin daÄŸÄ±lÄ±mÄ±nÄ± gÃ¶sterir.

### Notlar

- HistogramlarÄ±n her biri, veri kÃ¼mesindeki farklÄ± bir sayÄ±sal Ã¶zelliÄŸi temsil eder.
- X eksenindeki deÄŸerler, veri kÃ¼mesindeki Ã¶zelliklerin aralÄ±ÄŸÄ±nÄ± belirtirken; Y ekseni, bu deÄŸerlerin kaÃ§ kez tekrarlandÄ±ÄŸÄ±nÄ± (frekansÄ±nÄ±) gÃ¶sterir.
- Grafiklerde gÃ¶rÃ¼len daÄŸÄ±lÄ±mlar, veri kÃ¼mesinin genel yapÄ±sÄ±nÄ± anlamamÄ±za yardÄ±mcÄ± olur ve bu Ã¶zelliklerin Ã§eÅŸitli analizlerde nasÄ±l kullanÄ±labileceÄŸi konusunda bilgi verir.


```python
# ID ve gereksiz sÃ¼tunlarÄ± Ã§Ä±kararak korelasyon matrisini hesapla
corr_matrix = df.drop(columns=["id"]).corr().abs()

# Korelasyon matrisinin Ã¼st Ã¼Ã§genini al
upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# En yÃ¼ksek ve en dÃ¼ÅŸÃ¼k korelasyonlarÄ± bul
most_correlated = upper_triangle.stack().sort_values(ascending=False)

# En yÃ¼ksek 10 korelasyon
print("En yÃ¼ksek 10 korelasyon:")
print(most_correlated.head(10))

# En dÃ¼ÅŸÃ¼k 10 korelasyon
print("\nEn dÃ¼ÅŸÃ¼k 10 korelasyon:")
print(most_correlated.tail(10))

# TÃ¼m korelasyonlarÄ± gÃ¶ster
print("\nTÃ¼m korelasyonlar:")
print(most_correlated)
```
